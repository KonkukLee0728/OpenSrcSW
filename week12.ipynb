```python
import csv
f = open('2020data.csv', encoding='euc-kr')
data = csv.reader(f)
header = next(data)

max_div = 0
div_date = ''
max_temp = 0
max_date = ''
min_temp = 0
min_date = ''

for row in data:
    if row[4] == '':
        row[4] = 0
    row[4] = float(row[4])
    if max_temp < row[4]:
        max_temp = row[4]
        max_date = row[0]
        
    if row[3] == '':
        row[3] = 0
    row[3] = float(row[3])
    if min_temp > row[3]:
        min_temp = row[3]
        min_date = row[0]
        
    gap = row[4] - row[3]
    if max_div < gap:
        max_div = gap
        div_date = row[0]

f.close()
print(f"일교차가 가장 큰날\n날짜 {div_date}")
print(f"최저기온이 가장 낮은 날\n날짜 {min_date}")
print(f"최고기온이 가장 높은날\n날짜 {max_date}")
```

```python
import csv
f = open('202104card.csv')
data = csv.reader(f)
save = []
next(data)

max_user = 0


for row in data:
    if type(row) != type([]):
        break
    if row[4] == '':
        row[4] = 0
    if row[5] == '':
        row[5] = 0
    row[4] = row[4].replace(',', '')
    row[5] = row[5].replace(',', '')
    sum = int(row[4]) + int(row[5])
    row.append(sum)
    save.append(row)
    
save.sort(key=lambda x:x[6], reverse=True)
    

print(save)

        

f.close()

print("이용객이 가장 많은 역 순위")
for i in range(0, 3):
    print(f"{i+1}. {save[i][3]}, ({save[i][6]})")

print("이용객이 가장 적은 역 순위")
for i in range(-1, -4, -1):
     print(f"{-1*i}. {save[i][3]}, ({save[i][6]})")
        

    
```
