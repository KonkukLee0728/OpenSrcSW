```python
import csv

import matplotlib.pyplot as plt

f = open("seoul.csv", encoding='euc-kr')

data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        if row[0].split('.')[1] == '7' and row[0].split('.')[2] == '29' :
            result.append(float(row[-1]))
print(result)

plt.title('birthday high temp')
plt.plot(result)
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding = 'euc-kr')

data = csv.reader(f)
next(data)

jan = []
aug = []

for row in data:
    month = row[0].split('.')[1]
    if row[-1] != '':
        if month == '1':
            jan.append(float(row[-1]))
        if month == '8':
            aug.append(float(row[-1]))

plt.hist(jan, bins = 100 ,color = 'b', label = 'Jan')
plt.hist(aug, bins = 100 ,color = 'r', label = 'Aug')
plt.show()
```
